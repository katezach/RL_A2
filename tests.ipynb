```python
import os 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

dir = 'run_results'
# print(os.listdir(dir))

results = [np.load(os.path.join(dir, file)) for file in os.listdir(dir)]
result_name = [file.split('.npy')[0] for file in os.listdir(dir)]
print(result_name)
x_linspace = np.linspace(1, 2000, 2000)
print(x_linspace)

for (result, result_name) in zip(results, result_name):
    learning_rate = result_name.split('_')[2]
    gamma = result_name.split('_')[3]

    result = savgol_filter(result, 101, 3)
    plt.plot(x_linspace, result.mean(axis=0), label=f'γ={gamma}, α={learning_rate}$')
    plt.fill_between(x_linspace, result.mean(axis=0) - result.std(axis=0), result.mean(axis=0) + result.std(axis=0), alpha=0.3)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Actor Critic Experiments')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
```

```python
import os 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

dir = 'ac_baseline'
# print(os.listdir(dir))

results = [np.load(os.path.join(dir, file)) for file in os.listdir(dir)]
result_name = [file.split('.npy')[0] for file in os.listdir(dir)]
print(result_name)
x_linspace = np.linspace(1, 2000, 2000)
print(x_linspace)

for (result, result_name) in zip(results, result_name):
    learning_rate = result_name.split('_')[1]
    weight = result_name.split('_')[3]

    result = savgol_filter(result, 101, 3)
    plt.plot(x_linspace, result.mean(axis=0),label=f'α={learning_rate}, w={weight}')
    plt.fill_between(x_linspace, result.mean(axis=0) - result.std(axis=0), result.mean(axis=0) + result.std(axis=0), alpha=0.3)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Actor Critic Experiments with Baseline Subtraction')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.savefig('plots/ac_baseline.png', bbox_inches='tight')
```

```python
import os 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

dir = 'ac_bootstrap'
# print(os.listdir(dir))

results = [np.load(os.path.join(dir, file)) for file in os.listdir(dir)]
result_name = [file.split('.npy')[0] for file in os.listdir(dir)]
print(result_name)
x_linspace = np.linspace(1, 2000, 2000)
print(x_linspace)

for (result, result_name) in zip(results, result_name):
    learning_rate = result_name.split('_')[1]
    weight = result_name.split('_')[5]
    boot = result_name.split('_')[3]
    result = savgol_filter(result, 101, 3)
    plt.plot(x_linspace, result.mean(axis=0),label=f'α={learning_rate}, w={weight}, n={boot}')
    plt.fill_between(x_linspace, result.mean(axis=0) - result.std(axis=0), result.mean(axis=0) + result.std(axis=0), alpha=0.2)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Actor Critic Experiments with n-step Bootstrapping')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.savefig('plots/ac_boot_variance.png', bbox_inches='tight')
```
